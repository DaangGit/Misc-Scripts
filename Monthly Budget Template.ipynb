```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (9,6)
```

## Income
**The most important part of your budget. Everything else is deducted from your income to ideally build a zero sum budget.**
Contains items like
- Monthly fixed income
- Secondary sources of income
- Misc inflows

---


```python
income = ['Monthly Fixed Income', 'Secondary Income', 'Misc Income']
income_df = pd.DataFrame(index = income)
for i in income_df.index:
    income_df.loc[i, 'Amount'] = float(input(i+': '))
```

```python
plt.pie(income_df['Amount'], labels = np.where(income_df['Amount'] > 0, income_df.index, None))
plt.title('Income Breakdown')
plt.show()
```

## Expenses
Expenses will further be broken down into fixed expenses and discretionary expenses.

------

### Fixed Expenses
Accounts for essentials like:
- Food/Groceries
- Utilities
- Travel
- Rent
- Other Bills

And expenses that increase your financial stability:
- Insurance
- EMIs
- Credit Card Payments
- Loan Repayments


```python
fixed_expenses = ['Food', 'Utilities', 'Travel', 'Rent', 'Other Bills', 'Insurance', 'EMIs', 'Credit Cards','Loan Repayment']
fixed_exp_df = pd.DataFrame(index = fixed_expenses)
for i in fixed_exp_df.index:
    fixed_exp_df.loc[i,'Amount'] = float(input(i+': '))
```

```python
plt.pie(fixed_exp_df['Amount'], labels = np.where(fixed_exp_df['Amount'] > 0, fixed_exp_df.index, None))
plt.title('Fixed Expenses Breakdown')
plt.show()
```

### Discretionary Expenses
Stuff that's not exactly necessary but nice to have.

```python
disc_expenses = ['Shopping', 'Restaurants', 'Entertainment', 'Other stuff']
disc_exp_df = pd.DataFrame(index = disc_expenses)
for i in disc_exp_df.index:
    disc_exp_df.loc[i,'Amount'] = float(input(i+': '))
```

```python
plt.pie(disc_exp_df['Amount'], labels = np.where(disc_exp_df['Amount'] > 0, disc_exp_df.index, None))
plt.title('Discretionary Expenses Breakdown')
plt.show()
```

```python
expenses_df = pd.concat([fixed_exp_df, disc_exp_df])
plt.pie(expenses_df['Amount'], labels = np.where(expenses_df['Amount'] > 0, expenses_df.index, None))
plt.title('Total Expenses Breakdown')
plt.show()
```

## Savings
Do with these what you will.

```python
# Savings
savings = income_df['Amount'].sum() - expenses_df['Amount'].sum()
print(savings)
```
